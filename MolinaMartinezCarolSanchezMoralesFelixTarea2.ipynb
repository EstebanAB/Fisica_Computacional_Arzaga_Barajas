 ### Implemetación del pseudocódigo de un programa que calcula el producto de los números pares hasta un número determinado N. Si el número N es par, debe multiplicar hasta ese número, sino, debe multiplicar hasta N-1.

```julia
N=60                   #= Asignamos un valor determinado al número más grande con que se trabajará. Ej:5 =#
pp=1                     #= Asignamos al producto de los números pares (pp) el valor inicial de 1.=#
i=2
if(N%2==0)               #= Si se cumple que i%2==0 (un número es par si el residuo del número dividido entre 2 es cero) se entra en un ciclo while=#
    while (2<=i<=N)      #= El ciclo while correrá desde  que i= 2 hasta i=N.=#
        if (i%2==0)      #= Si i es par, se hará la siguiente instrucción: pp=pp*i. =#
            pp=pp*i
        end
        i +=1            #= Si i no es par, se avanzará al sigiente número entero. =#
    end
    print(pp)            #= Una vez finalizado el ciclo while, se imprimirá el valor final de pp. =#
else                     #= Para el caso en que N no sea par. =#
    while (2<=i<=N-1)    #= El ciclo while correrá desde  que i= 2 hasta i=N. =#
        if(i%2==0)       #= Si i es par, se hará la siguiente instrucción: pp=pp*i. =#
            pp=pp*i
        end
        i +=1            #= Si i no es par, se avanzará al sigiente número entero. =#
    end
    print(pp)            #= Una vez finalizado el ciclo while, se imprimirá el valor final de pp. =#
end

# H --- La idea no está mal, sin embargo, es posible reducir las líneas de tal suerte que quede todo en --- H #
# H --- un sólo ciclo, definiendo la condición de paro y la forma en cómo se ejecutará dicho cicho --- H #
```

```julia
function f(t,w)# Función que establece la ecuación diferencial
    return w-(1/2)*exp(t/2)*sin(5t) + 5exp(t/2)*cos(5t)
end
```

```julia
function eulerMethode(t0,N,h)
    t = zeros(1,N)
    w = zeros(1,N)
    t[1] = t0
    w[1] = 0
    println(t[1], " ", w[1])
    for i in 2:N
        t[i] = t[1] + i*h # se considera un paso de timpo costante
        w[i] = w[i-1] + h*f(t[i-1],w[i-1])
        println(t[i], " ", w[i])
    end
    (t,w)
end
```

```julia
(t,w)=eulerMethode(0,100,0.05) # Seran 100 intervalos de longitud 0.05, comenzando en un tiempo
```

### Runga Kutta

```julia
h=0.05
```

```julia
function F2(t,w)
    return f(t+h/2,w+(h/2)*f(t,w))
end
```

```julia
function F3(t,w)
    return f(t+h/2,w+(h/2)*f(t+h/2,w+(h/2)*f(t,w)))
end
```

```julia
function F4(t,w)
    return f(t+h,w+h*f(t+h/2,w+(h/2)*f(t+h/2,w+(h/2)*f(t,w))))
end
```

```julia
function f(t,w)
    return w-(1/2)*exp(t/2)*sin(5t) + 5exp(t/2)*cos(5t)
end
```

```julia
function RungaKutta(t0,N,h) 

W = zeros(1,N)
w = zeros(1,N)
t[1] = t0
W[1] = 0
    for k in 1:N-1
        t[k+1] = t[k] + h
        W[k+1]=W[k]+(h/6)*(f(t[k],w[k])+2(f(t[k]+h/2,w[k]+(h/2)*f(t[k],w[k])))+(f(t[k]+h/2,w[k]+(h/2)*f(t[k]+h/2,w[k]+(h/2)*f(t[k],w[k]))))+(f(t[k]+h,w[k]+h*f(t[k]+h/2,w[k]+(h/2)*f(t[k]+h/2,w[k]+(h/2)*f(t[k],w[k]))))))
        println(t[k], " ", W[k])
    end
    (t,W)
end
```

```julia
(t,W)=RungaKutta(0,100,0.05) 
```

```julia
ts = linspace(0,5,90)
ys = map(t -> e^(t/2)*sin(5t), ts)
plot(ts,ys,color="pink")
scatter(t,W, color = "red", s = 3)
```

La solución obtendida con el método de Runge-Kutta de cuarto orden se aproxima más a la solución exacta de la ecuación diferencial.
