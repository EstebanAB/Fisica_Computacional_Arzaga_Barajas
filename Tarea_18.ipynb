## Mapa logistico

```julia
using PyPlot
```

```julia
function mapa_logistico(r,x0)
    lista_x=[]
    for i in 1:length(r)
        x = x0
        for q in 1:100000
            x = r[i]*x*(1-x)
        end
        push!(lista_x,x)
    end
    xlabel(L"r")
    ylabel(L"x")
    title("Diagrama de bifurcaciones")
    plot(r,lista_x,color="b");
end;
```

```julia
mapa_logistico(0:0.0001:5,0.5555);
mapa_logistico(0:0.0001:5,0.7777);
mapa_logistico(0:0.0001:5,0.42222);
mapa_logistico(0:0.0001:5,0.33333);
mapa_logistico(0:0.0001:5,0.25555);
```
